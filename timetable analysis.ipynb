```python
Monday=(pd.read_excel('TimeTable.xlsx', skiprows=3, sheet_name='Monday'))#reading Monday data
Tuesday=(pd.read_excel('TimeTable.xlsx', skiprows=3, sheet_name='Tuesday'))#reading Monday data
Wednesday=(pd.read_excel('TimeTable.xlsx', skiprows=3, sheet_name='Wednesday'))#reading Monday data
Thursday=(pd.read_excel('TimeTable.xlsx', skiprows=3, sheet_name='Thursday'))#reading Monday data
Friday=(pd.read_excel('TimeTable.xlsx', skiprows=3, sheet_name='Friday'))#reading Monday data
```

```python
def Dataclass(org): #function to drop unnamed nan columns from all dataframes
    day=org.rename(columns=org.iloc[0]).drop(org.index[0])
    day.columns = day.columns.fillna('to_drop') #droping NaN named columns by renaming them todrop first
    day=day.drop('to_drop', axis = 1) #dropped
    return day

def classSeperator(org): #seperating classes as it has lab data too
    return (org.drop(org.index[36:]))
    
MondayClass=Dataclass(Monday)  
TuesdayClass=Dataclass(Tuesday)  
WednesdayClass=Dataclass(Wednesday)  
ThursdayClass=Dataclass(Thursday) 
FridayClass=Dataclass(Friday)  

MondayClassOnly= classSeperator(MondayClass) #MondayClass.drop(MondayClass.index[36:]) #separating class data from lab one at the end
TuesdayClassOnly=classSeperator(TuesdayClass)
WednesdayClassOnly=classSeperator(WednesdayClass)
ThursdayClassOnly=ThursdayClass.drop(ThursdayClass.index[35:])
FridayClassOnly=FridayClass.drop(FridayClass.index[33:])
```

```python
def LabData(org): #orginal data
    tempcopy=org.iloc[36:,:] #For extracting lab data only
    Lab=tempcopy.rename(columns=tempcopy.iloc[0]).drop(tempcopy.index[0]) #renaming
    Lab.columns = Lab.columns.fillna('to_drop')
    Lab=Lab.drop('to_drop', axis = 1)
    return Lab

#for thursday it had 35 rows
def LabData1(org): #orginal data
    tempcopy=org.iloc[35:,:] #For extracting lab data only
    Lab=tempcopy.rename(columns=tempcopy.iloc[0]).drop(tempcopy.index[0]) #renaming
    Lab.columns = Lab.columns.fillna('to_drop')
    Lab=Lab.drop('to_drop', axis = 1)
    return Lab
#for friday it had 33
def LabData2(org): #orginal data
    tempcopy=org.iloc[33:,:] #For extracting lab data only
    Lab=tempcopy.rename(columns=tempcopy.iloc[0]).drop(tempcopy.index[0]) #renaming
    Lab.columns = Lab.columns.fillna('to_drop')
    Lab=Lab.drop('to_drop', axis = 1)
    return Lab

def indexLab(day): #set lab name as index
    return day.set_index('Lab')

MondayLab=LabData(MondayClass)
MondayLab=indexLab(MondayLab)

TuesdayLab=LabData(TuesdayClass)
TuesdayLab=indexLab(TuesdayLab)

WednesdayLab=LabData(WednesdayClass)
WednesdayLab=indexLab(WednesdayLab)

ThursdayLab=LabData1(ThursdayClass)
ThursdayLab=indexLab(ThursdayLab)

FridayLab=LabData2(FridayClass)
FridayLab=indexLab(FridayLab)
```

```python
def checkClasses(subject,day):
    name=0
    col=[] 
    temp=[] 
    extra=[]
    if subject in day.values:
        s = (day == subject).any()
        a = s.index[s]
        k=day.columns.get_indexer(a) #col indexes where value true
    
        for i in range(0,len(k)):
            col.append(k[i])
        colL=len(col)
        for i in range(0,len(col)): #loop to find row indexes
            if col[i]==7:
                name="05:20-06:40 "
            elif col[i]==1:
                name="08:30-09:50"
            elif col[i]==2:
                name="10:00-11:20"
            elif col[i]==3:
                name="11:30-12:50"
            elif col[i]==4:
                name="01:00-02:20"
            elif col[i]==5:
                name="02:30-03:50"
            elif col[i]==6:
                name="03:55-05:15"
            elif col[i]==8:
                name="06:45-08:05"
            rows=day.index[day[name]==subject].tolist() #list of row indexes
            #print(rows)
            temp.append(rows) #if subject exist in more than 1 column temp list comes into play
  
        col.insert(0,0) #0 shows room column
        p=len(temp)
        pp=len(rows)
        if(colL>1): #if subject exists in more than 1 column
            for i in range(0,p):
                extra.append(temp[i][0])
            extra = [*set(extra)] #removing duplicates
            for i in range(0,len(extra)):
                 extra[i]-=1
        
            print(day.iloc[extra, col])
        else:
            for i in range(0,pp):
                extra.append(temp[0][i])
            for i in range(0,pp):
                extra[i]-=1
            print(day.iloc[extra, col])
    else:
        print("No classes")
def checkFriday(subject,day): #friday had different timings so had to writes its own function 
    name=0
    col=[]
    temp=[]
    extra=[]
    if subject in day.values:
    #print('yes')
        s = (day == subject).any()
        a = s.index[s]
        k=day.columns.get_indexer(a)
    
        for i in range(0,len(k)):
            col.append(k[i])
        colL=len(col)
        #print(col)
        for i in range(0,len(col)):
            if col[i]==7:
                name="05:20-06:40 "
            elif col[i]==1:
                name="08:30-09:50"
            elif col[i]==2:
                name="10:00-11:20"
            elif col[i]==3:
                name="11:30-12:50"
            elif col[i]==4:    
                #name="01:00-02:20"
                k=2
            elif col[i]==5:
                name="02:00-03:20"
            elif col[i]==6:
                name="03:30-04:50"
            elif col[i]==8:
                name="06:45-08:05"
            rows=day.index[day[name]==subject].tolist()
            #print(rows)
            temp.append(rows)
  
        col.insert(0,0)
        p=len(temp)
        pp=len(rows)
        if(colL>1):
            for i in range(0,p):
                for j in range(0,len(temp[i])):
                    extra.append(temp[i][j])
            extra = [*set(extra)] #removing duplicates
            print(extra)
            for i in range(0,len(extra)):
                 extra[i]-=1
            if(subject=="FYP / MS Thesis Evaluations"):
                t=day.replace(['Applied Programming (AI) Cancelled','FSM Graduate Classes'], 'Nill') #
                print(t.iloc[extra, col])
            else:
                print(day.iloc[extra, col])
        else:
            for i in range(0,pp):
                extra.append(temp[0][i])
            for i in range(0,pp):
                extra[i]-=1
            print(day.iloc[extra, col])
    else:
        print("No classes")

subject=input("Enter subject in right format 'subject[space](Department-Section)'")
print("MONDAY:")
checkClasses(subject,MondayClassOnly)
print("TUESDAY:")
checkClasses(subject,TuesdayClassOnly)
print("WEDNESDAY:")
checkClasses(subject,WednesdayClassOnly)
print("THURSDAY:")
checkClasses(subject,ThursdayClassOnly)
print("FRIDAY:")
checkFriday(subject,FridayClassOnly)       
```

```python
def printEmpty(day):
    rows_with_nan = [index for index, row in day.iterrows() if row.isnull().any()] #number of rows index with nan value
    for i in range(0,len(rows_with_nan)):
        rows_with_nan[i]-=1 #-1 to print
    col=[]    
    res = day.columns[day.isnull().any(0)]
    k=day.columns.get_indexer(res) #column numbers values

    M=day.fillna('Empty') #fill NaN with Empty
    for i in range(0,len(k)):
         col.append(k[i])
    col.pop(0)       
    for i in range(0,len(rows_with_nan)):
        for j in range(0,len(col)):
            if M.iloc[rows_with_nan[i], col[j]]=='Empty':
                L=M.iloc[rows_with_nan[i], [0,col[j]]]
                print(L.to_string())
print("-----MONDAY------")               
printEmpty(MondayClassOnly)
```
